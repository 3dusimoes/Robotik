```julia
using PyPlot
```

```julia
function single_neuron(w,bias,n,init)
    a = init
    o = zeros(n)
    for i = 1:n
        o[i] = tanh(a)
        a = bias + w * o[i]
    end
    o
end
```

```julia
n = collect(-5:0.01:5)
outn2 = zeros(100,length(n))
for s = 1:length(n)
    outn2[:,s] = single_neuron(-2,n[s],100,rand() * 0.02 - 0.01)
end
writecsv("/Users/zahedi/Desktop/single_n2.csv",outn2)


n = collect(-5:0.01:5)
outp2 = zeros(100,length(n)*2)
first = true
last = -rand() * 0.01
for s = 1:length(n)
    l = single_neuron(2,n[s],100,last)
    outp2[:,s] = l
    last = l[end]
end
first = true
last = -rand() * 0.01
for s = length(n):-1:1
    l = single_neuron(2,n[s],100,last)
    outp2[:,100+s] = l
    last = l[end]
end
writecsv("/Users/zahedi/Desktop/single_n2.csv",outp2)
```

```julia
a=outp2[end-1:end,:]
figure(figsize=(5*1.61,5))
xticks([])
yticks([])
ylim([-1.1,1.1])
plot(a[1,:]',".",color="blue")
plot(a[2,:]',".",color="blue")
savefig("/Users/zahedi/Desktop/single_p2.png")
```
