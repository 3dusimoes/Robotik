```julia
using PyPlot
using ODE
```

```julia
function f(t, v)
    (theta,thetadot)  = v
    L = 1
    g = 9.81
    [thetadot; -g/L * sin(theta)]
end
figure(figsize=(12,6)); ylim([-2pi,2pi]); xlim([-4pi,4pi])
time = 0:0.01:10;
yinit = [[-5*pi 1.0];[5*pi -1.0];[0 0.2];[0 1.5];[0 3.5];[0 4.5]]
y0 = zeros(2)
for i = 1:size(yinit)[1]
    y0[:] = yinit[i,:]
    t,y = ode45(f, y0, time);
    plot([v[1] for v in y],[v[2] for v in y])
end

savefig("/Users/zahedi/Desktop/fig.png")
```
